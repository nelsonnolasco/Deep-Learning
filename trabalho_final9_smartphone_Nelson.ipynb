<h1><center>Reconhecimento de atividade humana pelo smartphone</center></h1>

Nesta atividade iremos usar informações capturadas pelo acelerômetro e giroscópio dos celulares para dizer qual das 6 atividades o usuário está executando: 

* Andando
* Subindo escadas
* Descendo escadas
* Sentando
* Levantando
* Deitando

<h3>Sobre o dataset</h3>
As experiências foram realizadas com um grupo de 30 voluntários com idade entre 19 e 48 anos. Cada pessoa realizou seis atividades (WALKING, WALKING_UPSTAIRS, WALKING_DOWNSTAIRS, SITTING, STANDING, LAYING) usando um smartphone (Samsung Galaxy S II) na cintura. Usando seu acelerômetro e giroscópio embutido, capturamos a aceleração linear 3-axial e a velocidade angular 3-axial a uma taxa constante de 50Hz. As experiências foram gravadas em vídeo para rotular os dados manualmente. O conjunto de dados obtido foi dividido aleatoriamente em dois conjuntos, onde 70% dos voluntários foram selecionados para gerar os dados de treinamento e 30% os dados de teste.

Temos 561 features, com valores capturados pelos sensores do celular. A penúltima coluna, _subject_,  indentifica o indivíduo que executou a ação. A última coluna diz qual a atividade foi executada (é isso que nossa rede deve aprender).

# Preparando os dados

A única coisa que precisamos fazer é separar os labels do restante dos dados. Tanto do treino, quanto do teste. Além de separar, vamos converter as categorias para números, ou seja, ao invés de prever "WALKING", nossa rede precisa prever o número 1, e assim por diante. Além disso, vamos excluir a coluna subject, que não nos interessa para fazer as previsões.

```
x_train = df_train.copy()
x_train.Activity = pd.Categorical(x_train.Activity)
x_train['Activity'] = x_train.Activity.cat.codes
y_train = x_train['Activity'].copy()
x_train.drop(['subject', 'Activity'], axis=1, inplace=True)

x_test = df_test.copy()
x_test.Activity = pd.Categorical(x_test.Activity)
x_test['Activity'] = x_test.Activity.cat.codes
y_test = x_test['Activity'].copy()
x_test.drop(['subject', 'Activity'], axis=1, inplace=True)
```

# Criando o modelo

Seu trabalho é estruturar uma rede neural profunda para esta tarefa. Para isso adicione dense layers ao modelo abaixo (função _model.add()_). Coloque quantas camadas você achar necessário, especificando a quantidade de neurônios e qual função de ativação.


Lembre-se de especificar a camada de entrada para as 561 features do nosso dataset.


Lembre-se que a última camada (camada de saída) deve estar de acordo com o nosso problema: Uma tarefa  de classificação onde temos 6 classes. Isso influencia no número de neurônios e qual ativação usar na última camada.

<br>

Para vocês terem uma base, no momento do teste lá embaixo, seu modelo deve obter pelo menos 90% de acurácia. Se isso não acontecer, aumente o número de camadas e/ou de neurônios.

```
model = tf.keras.Sequential()


model.add(tf.keras.layers.Input(561))
model.add(tf.keras.layers.Dense(50, activation='relu'))
model.add(tf.keras.layers.Dense(6, activation='sigmoid'))

model.compile(optimizer='adam', loss=tf.keras.losses.sparse_categorical_crossentropy, metrics='accuracy')
```

# Treinando

```
model.fit(x_train, y_train, epochs=100, batch_size=500)
```

# Testando

```
model.evaluate(x_test, y_test)
```
